
# Bài tập 4: Vẽ biểu đồ top 20 bài hát có lượng play time cao nhất ở Mỹ (Region == "us")
Lưu ý: Play time = Streams * duration

```python
#count stream
total_stream_df = data_df[data_df.Region == "us"].groupby(["Track Name"])["Streams"].agg(["sum"])
total_stream_df.rename(columns={"sum":"total_streams"},inplace=True)

playtime_df = feature_df.set_index(["name"]).join(total_stream_df,how="left")
playtime_df["play_time"] = (playtime_df["duration"].values * playtime_df["total_streams"].values)/86400
playtime_df
```

```python
top20_df = playtime_df.nlargest(20,["play_time"])[["play_time"]]

play_time_ax = top20_df.sort_values(by=["play_time"],ascending=True).plot.barh(use_index=True, y='play_time')
```

# Bài tập 7: Sử dụng PCA để biểu diễn các bài hát bằng biểu đồ scatter (dùng dữ liệu featuresdf). So sánh với việc sử dụng TSNE để giảm chiều (giải thích các tham số trong TSNE)

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

## Sử dụng PCA:

```python
# Chuẩn hoá dữ liệu
numeric_features = feature_df.select_dtypes(include=[np.number])
scaled_features = StandardScaler().fit_transform(numeric_features)

# Áp dụng PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Tạo DataFrame cho kết quả PCA
pca_df = pd.DataFrame(data = pca_result, columns = ['PCA1', 'PCA2'])

# Ghép label (nhãn tên bài hát) vào dữ liệu
pca_df = pd.concat([pca_df, feature_df[['name', 'artists']]], axis = 1)
pca_df
```

```python
# Vẽ biểu đồ scatter
plt.figure(figsize=(10,8))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'])

for i, txt in enumerate(pca_df['name']):
    plt.annotate(txt, (pca_df['PCA1'][i], pca_df['PCA2'][i]))

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')
plt.grid()
plt.show()
```

## So sánh với việc sử dụng TSNE để giảm chiều:

##### PCA (Principal Component Analysis):
- Tuyến tính: PCA là một phương pháp tuyến tính, nghĩa là nó không thể phát hiện mối quan hệ phi tuyến trong dữ liệu.
- Biến đổi tối ưu hóa phương sai: PCA giảm chiều dữ liệu bằng cách tìm ra các "thành phần chính" mới mà trên đó dữ liệu có biến đổi lớn nhất (phương sai lớn nhất).
- Tính toán nhanh: PCA thường nhanh hơn t-SNE và phù hợp để xử lý dữ liệu có số chiều rất lớn.
- Kết quả nhất quán: PCA cho ra kết quả nhất quán với cùng một bộ dữ liệu.

##### t-SNE (t-distributed Stochastic Neighbor Embedding):
- Phi tuyến: t-SNE là một phương pháp phi tuyến, có khả năng phát hiện các mối quan hệ phức tạp và cấu trúc trong dữ liệu nhiều chiều.
- Giữ nguyên cấu trúc cục bộ: t-SNE tối ưu hóa không gian sao cho các điểm gần nhau trong không gian nhiều chiều vẫn gần nhau trong không gian giảm chiều.
- Tính toán chậm và tốn tài nguyên hơn: t-SNE thường mất nhiều thời gian hơn để tính toán, đặc biệt với các bộ dữ liệu lớn.
- Kết quả không nhất quán: t-SNE có thể cho ra kết quả khác nhau khi chạy nhiều lần trên cùng một bộ dữ liệu do sử dụng thuật toán ngẫu nhiên.

##### Các tham số quan trọng trong t-SNE:
- n_components: Số chiều của không gian đầu ra.
- perplexity: Một tham số liên quan đến số lượng hàng xóm lân cận; giá trị thông dụng từ 5 đến 50.
- early_exaggeration: Ảnh hưởng đến sự tách biệt giữa các cụm dữ liệu; giá trị cao hơn có thể giúp cụm rõ ràng hơn.
- learning_rate: Ảnh hưởng đến tốc độ học của thuật toán; giá trị thông thường từ 10 đến 1000.
- n_iter: Số vòng lặp tối đa cho quá trình tối ưu hóa.
- n_iter_without_progress: Số vòng lặp mà sau đó, nếu không có cải thiện, quá trình tối ưu hóa sẽ dừng lại.
- random_state: Đặt giá trị cố định cho seed của quá trình ngẫu nhiên, có thể giúp tái tạo lại kết quả.
